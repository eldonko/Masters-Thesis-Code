```python
import matplotlib.pyplot as plt
import numpy as np
from numpy.linalg import pinv
from numpy.polynomial.polynomial import polyval
```

```python
v = np.array([[.25, .5, 1], [1, 1, 0], [2, 1, 0]])
w = np.array([[1], [0], [1]])

a = np.flip((pinv(v) @ w).squeeze())
print(a)
```

```python
x = np.array([[0.6, 1, 0], [0.3**2, 0.3, 1], [1, 1, 1]])
y = np.array([[0], [1], [1.4]])

b = np.flip((pinv(x) @ y).squeeze())
print(b)
```

```python
x = np.arange(0, 101)/100
```

```python
p = polyval(x, a)
q = polyval(x, b)
```

```python
plt.plot(x, p)
plt.plot(x, q)
```
