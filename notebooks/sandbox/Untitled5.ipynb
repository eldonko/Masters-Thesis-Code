```python
import matplotlib.pyplot as plt
import torch
```

```python
a = torch.tensor([[0, 0],
                  [1, 1],
                  [2, 2], 
                  [3, 3],
                  [4, 4]])

b = torch.tensor([[0, 1],
                  [1, 1.5],
                  [2, 2], 
                  [3, 2.5],
                  [4, 3]])
```

```python
plt.plot(a[:, 0], a[:, 1])
plt.plot(b[:, 0], b[:, 1])
```
