```python
import numpy as np
import torch
import torch.nn as nn
from torch.autograd import grad
import matplotlib.pyplot as plt
```

```python
seq = nn.Sequential(
    nn.Linear(1, 128),
    nn.ReLU(),
    nn.Linear(128, 128),
    nn.ReLU(),
    nn.Linear(128, 100)
    #nn.ReLU()
    )
```

```python
inp = torch.tensor([4.])
out = seq(inp)

print(out)
out_ad = torch.where(out > 0, 1., 0.)
print(out_ad)
```

```python
gt = torch.tensor([0., 1., 0., 0.])
```

```python
def ineq_loss(output, target):
    #return torch.tensor([(output == target).sum().float()], requires_grad=True)
    return output + target
```

```python
loss = ineq_loss(out, gt)
print(loss)
```

```python
loss_fn = nn.MultiLabelMarginLoss()
```

```python
out = torch.tensor([0., 0.9, 0., 0.])
gt = torch.zeros(100, dtype=torch.long)
gt[0:9] = torch.tensor([1, 4, 8, 13, 25, 38, 61, 79, -1])
```

```python
optimizer = torch.optim.Adam(seq.parameters(), lr=.1)

for i in range(100):
    inp = torch.tensor([4.])
    out = seq(inp)
    
    loss = loss_fn(out, gt)
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
    print(loss)
    #print(out)
    #print(gt)
    print()
print(out)
print(gt)
```
