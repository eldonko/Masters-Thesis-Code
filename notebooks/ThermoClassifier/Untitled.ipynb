```python
import numpy as np
from sklearn import datasets
from sklearn.multiclass import OneVsOneClassifier
from sklearn.svm import LinearSVC
#X, y = datasets.load_iris(return_X_y=True)
```

```python
length = 100
X, y = np.zeros([length, 4]), np.zeros([length, ])


for i in range(length):
    b = np.random.randint(1, 77)
    
    X[i] = np.random.randint((b - 1) * 10, b * 10, (4,))
    y[i] = b
        
out = OneVsOneClassifier(LinearSVC(random_state=0, max_iter=1000000)).fit(X, y).predict(X)
```

```python
print(*zip(out, y))

print('\naccuracy: ', (out == y).sum()/len(y))
```
